```python
import pandas as pd
import numpy as np
import os
```

```python
ST1 = pd.read_excel("../../../data/original/zhang/ST1.xlsx", sheet_name=0)
```

<b>ST1 

Table S1 contains list of peptides for each experiment in a weird format, I need to extract list of peptides for each experiment separately and them join them in a nice way

```python
ST1_Exp1 = ST1.iloc[:100, 0:3]
ST1_Exp1 = ST1_Exp1.rename(columns=ST1_Exp1.iloc[1])
ST1_Exp1 = ST1_Exp1[2:98].reset_index(drop=True)
ST1_Exp1["Experiment"] = "1"
```

```python
ST1_Exp1
```

```python
ST1_Exp2 = ST1.iloc[:100, 4:7]
ST1_Exp2 = ST1_Exp2.rename(columns=ST1_Exp2.iloc[1])
ST1_Exp2 = ST1_Exp2[2:98].reset_index(drop=True)
ST1_Exp2["Experiment"] = "2"
```

```python
ST1_Exp2
```

```python
ST1_Exp3 = ST1.iloc[:45, 8:11]
ST1_Exp3 = ST1_Exp3.rename(columns=ST1_Exp3.iloc[1])
ST1_Exp3 = ST1_Exp3[2:].reset_index(drop=True)
ST1_Exp3["Experiment"] = "3"
```

```python
ST1_Exp3
```

```python
ST1_Exp4 = ST1.iloc[:45, 12:15]
ST1_Exp4 = ST1_Exp4.rename(columns=ST1_Exp4.iloc[1])
ST1_Exp4 = ST1_Exp4[2:].reset_index(drop=True)
ST1_Exp4["Experiment"] = "4"
```

```python
ST1_Exp4
```

```python
ST1_Exp5 = ST1.iloc[:, 16:19]
ST1_Exp5 = ST1_Exp5.rename(columns=ST1_Exp5.iloc[1])
ST1_Exp5 = ST1_Exp5[2:].reset_index(drop=True)
ST1_Exp5["Experiment"] = "5"
```

```python
ST1_Exp5
```

```python
ST1_Exp6 = ST1.iloc[:,20:23]
ST1_Exp6 = ST1_Exp6.rename(columns=ST1_Exp6.iloc[1])
ST1_Exp6 = ST1_Exp6[2:].reset_index(drop=True)
ST1_Exp6["Experiment"] = "6"
```

```python
ST1_Exp6
```

```python
frames = [ST1_Exp1, ST1_Exp2, ST1_Exp3, ST1_Exp4, ST1_Exp5, ST1_Exp6]
result = pd.concat(frames).reset_index(drop=True)
result["HLA_type"] = "A02:01"
```

```python
result = result.rename(columns = {"Peptide Name": "peptide_name", "Sequence": "peptide_seq", "Tetramer Fluorescence": "tetramer_fluorescence", "Experiment": "no_experiment"})
```
