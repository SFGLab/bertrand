<h3> VDJ db

```python
vdj_db = pd.read_csv("../../../data/original/VDJdb_all.csv.gz", sep="\t", encoding='latin1')
```

```python
vdj_db
```

```python
vdj_db.columns
```

```python
vdj_db['Gene'].value_counts()
```

```python
vdj_db.reset_index(inplace=True)
```

```python
remaining_cols = [col for col in vdj_db.columns if col not in ['Gene', 'CDR3', 'V', 'J']]
remaining_cols
```

```python
vdj_db_new = vdj_db.pivot(index='index', columns='Gene', values=['CDR3','J','V'])
```

```python
vdj_db_new.columns  = vdj_db_new.columns.droplevel(1)
```

```python
vdj_db_new.columns
```

```python
vdj_db_new.columns = ['CDR3a', 'CDR3b', "TRAJ", 'TRBJ', 'TRAV', "TRBV"]
```

```python
vdj_final = pd.concat([vdj_db_new,vdj_db[remaining_cols]],axis=1)
```

```python
vdj_final
```

```python
vdj_final = vdj_final.drop(columns=['index','MHC B'])
```

```python
vdj_final.columns
```

```python
vdj_final = vdj_final.rename(columns={'complex.id':'complex_id', 'Species': 'tcr_source_organism', 'MHC A':'HLA_type', 'MHC class': 'MHC_class', 'Epitope':'peptide_seq', 'Epitope gene':'peptide_gene', 'Epitope species':'peptide_source', 'Reference':'tcr_reference', 'Method':'method', 'Meta':'meta', 'Score':'vdj_score', "CDR3fix":"vdj_CDR3fix" })
```

```python
vdj_final['tcr_source_organism'] = "human"
```

```python
vdj_final['HLA_type'].unique()
```

```python
vdj_final['HLA_type'] = vdj_final['HLA_type'].str.replace("*", "")
vdj_final['HLA_type'] = vdj_final['HLA_type'].str.replace("HLA-", "")
```

```python
vdj_final['MHC_class'] = vdj_final["MHC_class"].str.replace("MHCI", "I")
vdj_final['MHC_class'] = vdj_final["MHC_class"].str.replace("MHCII", "II")
```

```python
vdj_final["TRAV"] = vdj_final["TRAV"].str.replace("TRAV", "")
vdj_final["TRBV"] = vdj_final["TRBV"].str.replace("TRBV", "")
vdj_final["TRAJ"] = vdj_final["TRAJ"].str.replace("TRAJ", "")
vdj_final["TRBJ"] = vdj_final["TRBJ"].str.replace("TRBJ", "")
```

Look what method contains in example

```python
vdj_final["method"][169]
```

```python
experiment_type_dicts = vdj_final.method.apply(lambda dictionary: json.loads(dictionary))
```

```python
vdj_final["experiment_type"] = experiment_type_dicts.apply(lambda d: d['identification'])
```

```python
vdj_final["single_cell"] = experiment_type_dicts.apply(lambda d: d['singlecell'])
```

```python
vdj_final["sequencing"] = experiment_type_dicts.apply(lambda d: d['sequencing'])
```

```python
vdj_final["verification"] = experiment_type_dicts.apply(lambda d: d['verification'])
```

```python
meta_dicts = vdj_final.meta.apply(lambda dictionary: json.loads(dictionary))
```

```python
vdj_final["tissue"] = meta_dicts.apply(lambda d: d['tissue'])
```

```python
vdj_final["verification"] = vdj_final["verification"].str.strip()
```

```python
vdj_final["verification"].apply(lambda x: len(x)==0)
```

```python
vdj_final["experiment_type"] = vdj_final["experiment_type"] + "," +  vdj_final["verification"]
```

```python
vdj_final["experiment_type"] = vdj_final["experiment_type"].str.rstrip(',')
```

```python
vdj_final = vdj_final.drop(columns=["method", "meta", "verification"])
```

```python
column_names = ["peptide_seq", "HLA_type", "CDR3a", "CDR3b", "peptide_source", "peptide_gene", "MHC_class", "TRAV", "TRBV","TRAJ", "TRBJ", "tcr_source_organism", "tissue", "experiment_type", "single_cell", "sequencing", 'tcr_reference', 'vdj_CDR3fix', "vdj_score"]

vdj_final = vdj_final.reindex(columns=column_names)
```

```python
vdj_final = vdj_final.drop_duplicates()
```

```python
vdj_final.to_csv('../../../data/processed/vdj_final.csv.gz', index=False)
```
