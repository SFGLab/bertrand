```python
import numpy as np
import pandas as pd
import os
```

```python
soon = pd.read_csv("../../../data/original/soon_original.csv.gz")
```

```python
soon
```

```python
remaining_cols = [col for col in soon.columns if col not in ['TCR chain', 'CDR3 sequence', 'V gene', 'D gene', 'J gene']]
remaining_cols
```

```python
soon.reset_index(inplace=True)
```

```python
soon_gene = soon.pivot(index='index', columns = 'TCR chain', values = ['CDR3 sequence', 'V gene', 'D gene', 'J gene'])
```

```python
soon_gene
```

```python
soon_gene.columns  = soon_gene.columns.droplevel(1)
```

```python
soon_gene.columns = ['CDR3a', 'CDR3b', "TRAV", 'TRBV', 'TRAD', "TRBD", "TRAJ", "TRBJ"]
```

```python
soon_gene = soon_gene.drop(columns="TRAD")
```

```python
soon_final = pd.concat([soon_gene,soon[remaining_cols]],axis=1)
```

```python
soon_final
```

```python
soon_final = soon_final.drop(columns=["Frequency %", "Patient/donor"])
```

```python
soon_final = soon_final.rename(columns={"a":"CDR3a", "b":"CDR3b", "T cell epitopes": "peptide_seq"})
```

```python
soon_final
```

```python
soon_final['peptide_seq'][0:10] = "SLFWNEPAI"
soon_final['peptide_seq'][10:] = "LLWNTVWNM"
```

```python
soon_final["TRAV"] = soon_final["TRAV"].str.replace("TCRAV", "")
soon_final["TRBV"] = soon_final["TRBV"].str.replace("TCRBV", "")
soon_final["TRBD"] = soon_final["TRBD"].str.replace("TCRBD", "")
soon_final["TRAJ"] = soon_final["TRAJ"].str.replace("TCRAJ", "")
soon_final["TRBJ"] = soon_final["TRBJ"].str.replace("TCRBJ", "")
```

```python
soon_final["TRBD"] = soon_final["TRBD"].str.replace("unresolved", "")
```

```python
soon_final["HLA_type"] = "A02:01"
soon_final["peptide_source"] = "HEV"
soon_final["tcr_source_organism"] = "human"
soon_final["experiment_type"] = "dextramer-sort"
soon_final["tissue"] = "PBMC"
soon_final["dataset"] = "Soon"
```

```python
soon_final.columns
```

```python
soon_final = soon_final[["peptide_seq", "HLA_type", "CDR3a", "CDR3b", "TRAV", "TRAJ", "TRBV", "TRBJ", "TRBD", "peptide_source", "tcr_source_organism", "experiment_type", "tissue", "dataset"]]
```

```python
soon_final.to_csv("../../../data/processed/soon_input.csv.gz", index=False)
```
