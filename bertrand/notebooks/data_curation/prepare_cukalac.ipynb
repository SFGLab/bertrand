```python
import pandas as pd
import os
```

```python
NP66 = pd.read_csv("../../../data/original/cucalac_1.csv.gz")
PA224 = pd.read_csv("../../../data/original/cucalac_2.csv.gz")
F262 = pd.read_csv("../../../data/original/cucalac_3.csv.gz")
```

```python
NP66.head()
```

```python
NP66.columns = ["TRBV", "CDR3b", "TRBJ", "Length", "TRAV", "TRAJ"]
```

```python
NP66[["TRAV", "CDR3a"]] = NP66['TRAV'].str.split(" ", n=1, expand=True)
```

```python
NP66["peptide_seq"] = "ASNENMETM"
```

```python
PA224.head()
```

```python
PA224.columns = ["TRBV", "CDR3b", "TRBJ", "Length", "TRAV", "CDR3a", "TRAJ"]
```

```python
PA224["peptide_seq"] = "SSLENFRAYV"
```

```python
F262.head()
```

```python
F262.columns = ["TRBV", "CDR3b", "TRBJ", "Length", "TRAV", "CDR3a", "TRAJ"]
```

```python
F262["peptide_seq"] = "LSLRNPILV"
```

```python
cukalac = pd.concat([NP66, PA224, F262], axis=0)
```

```python
cukalac = cukalac.drop(columns="Length")
```

```python
cukalac
```

```python
cukalac["peptide_source"] = "InfluenzaA"
cukalac["tcr_source_organism"]= "C57BL/6J mice"
cukalac["MHC_type"] = "H-2Db"
cukalac["tissue"] = "spleen"
cukalac["experiment_type"] = "RT-PCR, tetramer-sort"
cukalac["dataset"] = "Cukalac"
```

```python
cukalac
```

```python
cukalac = cukalac[["peptide_seq", "MHC_type", "CDR3a", "CDR3b", "TRAV", "TRAJ", "TRBV", "TRBJ", "peptide_source", "tcr_source_organism", "tissue", "experiment_type", "dataset"]]
```

```python
cukalac.to_csv("../../../data/processed/cukalac_input.csv.gz", index=False)
```
