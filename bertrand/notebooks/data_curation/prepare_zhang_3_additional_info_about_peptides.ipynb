<h3>ST4 - Additional info about peptides from experiment 3 and experiment 4

```python
ST4 = pd.read_excel("../../../data/original/zhang/ST4.xlsx", sheet_name=0)
```

```python
ST4 = ST4.rename(columns=ST4.iloc[0])
ST4 = ST4[1:].reset_index(drop=True)
```

```python
ST4
```

```python
ST4.columns
```

```python
ST4 = ST4.rename(columns={'Wildtype Name':'wildtype_name', 'Mutant Name':'mut_name', 'Wildtype amino acid':'wt_aa',
       'Mutant amino acid':'mut_aa', 'Position of mutation in peptide':'mut_pos',
       'Wildtype peptide':'peptide_wt', 'Wildtype HLA-A2 Binding  NetMHC 4.0 (nM)':'wt_peptide_binding_score',
       'Mutant peptide':'peptide_mut', 'Mutant HLA-A2 Binding NetMHC 4.0 (nM)':'mut_peptide_binding_score', 'Source':'peptide_reference'})
```

Merging wiltype and peptide name as one column - peptide name. The same with peptide sequence. Adding a flag.

```python
remaining_cols = [col for col in ST4.columns if col not in ['wildtype_name', 'mut_name', 'peptide_wt', 'peptide_mut']]
remaining_cols
```

```python
ST4_name = ST4.melt(id_vars=remaining_cols, value_vars=['wildtype_name', 'mut_name'], value_name='peptide_name')
```

```python
ST4_seq = ST4.melt(id_vars=remaining_cols, value_vars=['peptide_wt', 'peptide_mut'], value_name='peptide_seq')
```

```python
ST4_name.sample(5)
```

```python
mask = ST4_name['variable'].str.contains(r'wildtype')
ST4_name.loc[mask, 'wildtype'] = 'yes'
ST4_name.loc[~mask, 'wildtype'] = 'no'
```

```python
ST4_final = pd.concat([ST4_name,ST4_seq['peptide_seq']], axis=1)
ST4_final
```

```python
ST4_final.drop(columns=["variable"],inplace=True)
```

```python
cols_to_use = ["peptide_reference", "peptide_name", "wildtype", "peptide_seq"]
```

```python
ST4_final=ST4_final[cols_to_use]
```

Peptide source - cancer type

```python
mask = ST4_final['peptide_reference'].str.contains(r'Cohen')
ST4_final.loc[mask, 'peptide_source'] = 'SKCM'
```

```python
mask = ST4_final['peptide_reference'].str.contains(r'Strønen')
ST4_final.loc[mask, 'peptide_source'] = 'SKCM'
```

```python
mask = ST4_final['peptide_reference'].str.contains(r'Rajasagi')
ST4_final.loc[mask, 'peptide_source'] = 'CLL'
```

```python
mask = ST4_final['peptide_reference'].str.contains(r'Carreno')
ST4_final.loc[mask, 'peptide_source'] = 'melanoma'
```

```python
ST4_final.to_csv("../../../data/processed/zhang/ST4_final.csv.gz")
```
